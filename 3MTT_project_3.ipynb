```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: set plot style
sns.set(style='whitegrid')
```

```python
# Load the Titanic dataset
df = pd.read_excel('/content/drive/MyDrive/3mtt python/TITANIC.xlsx')

# Preview the dataset
print(df.head())
```

```python
# Check dimensions
print("Shape:", df.shape)

# Check column info
print(df.info())

# Check basic statistics
print(df.describe(include='all'))
```

Question with aswers

1. What is the survival rate of passengers?

```python
survival_rate = df['survived'].mean()
print("Survival Rate: {:.2%}".format(survival_rate))
```

2. What is the gender distribution of passengers?

```python
print(df['sex'].value_counts())
sns.countplot(x='sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

3. How does survival rate differ by class?

```python
sns.barplot(x='pclass', y='survived', data=df)
plt.title("Survival Rate by Class")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()
```

4. What is the distribution of passenger ages?

```python
sns.histplot(df['age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

5. How many passengers embarked from each location?

```python
print(df['embarked'].value_counts())
sns.countplot(x='embarked', data=df)
plt.title("Passengers per Embarkation Point")
plt.xlabel("Embarkation Port")
plt.show()
```

6. What is the average fare paid per class?

```python
avg_fare_per_class = df.groupby('pclass')['fare'].mean()
print(avg_fare_per_class)
sns.barplot(x='pclass', y='fare', data=df)
plt.title("Average Fare per Class")
plt.ylabel("Average Fare")
plt.xlabel("Class")
plt.show()
```

7. How does gender affect survival rates?

```python
sns.barplot(x='sex', y='survived', data=df)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

8. What is the correlation between fare and survival?

```python
sns.scatterplot(x='fare', y='survived', data=df)
plt.title("Fare vs. Survival")
plt.xlabel("Fare")
plt.ylabel("Survival")
plt.show()
```

9. How does the number of siblings/spouses aboard affect survival?

```python
sns.barplot(x='sibsp', y='survived', data=df)
plt.title("Survival Rate by Siblings/Spouses Aboard")
plt.xlabel("Number of Siblings/Spouses")
plt.ylabel("Survival Rate")
plt.show()
```

10. What is the age distribution by class?

```python
sns.boxplot(x='pclass', y='age', data=df)
plt.title("Age Distribution by Class")
plt.xlabel("Class")
plt.ylabel("Age")
plt.show()
```

11. What is the survival rate by embarkation point?

```python
sns.barplot(x='embarked', y='survived', data=df)
plt.title("Survival Rate by Embarkation Point")
plt.xlabel("Embarked From")
plt.ylabel("Survival Rate")
plt.show()
```

12. What is the impact of having parents/children aboard on survival?

```python
sns.barplot(x='parch', y='survived', data=df)
plt.title("Survival Rate by Parents/Children Aboard")
plt.xlabel("Number of Parents/Children")
plt.ylabel("Survival Rate")
plt.show()
```
